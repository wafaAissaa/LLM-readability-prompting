```python
import pandas as pd
from sklearn.metrics import classification_report
import nltk

nltk.download('punkt')
from nltk.tokenize import word_tokenize

# Sample input DataFrame: df
# df has columns: 'Text', 'super_annot', 'AnnotatedTerm', 'TermLabel', 'Characters'

# Separate annotations for Annot1 and Annot2
annot1 = annotations[annotations['super_annot'] == 'Annot1']
annot2 = annotations[annotations['super_annot'] == 'Annot2']
annot3 = annotations[annotations['super_annot'] == 'Annot3']

# Group annotations by Text (assuming Text is unique per row)
texts = annotations['Text'].unique()

# Store token-level labels for both annotators
all_tokens = []
all_labels_annot1 = []
all_labels_annot2 = []
all_labels_annot3 = []

import nltk
from nltk.tokenize import word_tokenize, WhitespaceTokenizer

def label_tokens(text, annotations, overlap_threshold=0.5):
    tokens = word_tokenize(text)
    token_spans = list(WhitespaceTokenizer().span_tokenize(text))
    labels = ['O'] * len(tokens)

    for _, row in annotations.iterrows():
        start, end = map(int, row['Characters'].split('-'))
        term_label = row['TermLabel']

        for i, (tok_start, tok_end) in enumerate(token_spans):
            # Compute token-annotation overlap
            overlap_start = max(start, tok_start)
            overlap_end = min(end, tok_end)
            overlap_len = max(0, overlap_end - overlap_start)
            token_len = tok_end - tok_start

            # Condition 1: full containment
            if tok_start >= start and tok_end <= end:
                labels[i] = term_label
            # Condition 2: significant overlap
            elif overlap_len / token_len >= overlap_threshold:
                labels[i] = term_label
                """print(row["AnnotatedTerm"])
                print(f"[WARNING] Partial overlap detected:\n"
                      f"  → Token: '{text[tok_start:tok_end]}' [{tok_start}-{tok_end}]\n"
                      f"  → Annotated: '{text[start:end]}' [{start}-{end}] with label '{term_label}'\n"
                      f"  → Overlap ratio: {overlap_len / token_len:.2f}\n")"""
    return tokens, labels


# Process each text
for text in texts:
    tokens = word_tokenize(text)

    a1_annots = annot1[annot1['Text'] == text]
    a2_annots = annot2[annot2['Text'] == text]
    a3_annots = annot3[annot3['Text'] == text]

    _, labels1 = label_tokens(text, a1_annots)
    _, labels2 = label_tokens(text, a2_annots)
    _, labels3 = label_tokens(text, a3_annots)



    # Check length match
    assert len(labels1) == len(labels2) == len(tokens), f"Length mismatch in text: {text}"

    all_tokens.extend(tokens)
    all_labels_annot1.extend(labels1)
    all_labels_annot2.extend(labels2)
    all_labels_annot3.extend(labels3)

# Compute token-level F1
print("annot1 vs annot2")
print(classification_report(all_labels_annot1, all_labels_annot1, digits=4))

print("annot1 vs annot3")
print(classification_report(all_labels_annot2, all_labels_annot2, digits=4))

print("annot2 vs annot3")
print(classification_report(all_labels_annot3, all_labels_annot3, digits=4))
```

```python
marco_binaire = 0
micro_binaire = 0
macro_f1 = f1_score(all_labels_annot1, all_labels_annot2, average='macro')
marco_binaire += macro_f1
micro_f1 = f1_score(all_labels_annot1, all_labels_annot2, average='micro')
micro_binaire += micro_f1
print("Macro F1-score: annot1 vs annot2", round(macro_f1, 4))
print("Micro F1-score: annot1 vs annot2", round(micro_f1, 4))

macro_f1 = f1_score(all_labels_annot2, all_labels_annot3, average='macro')
marco_binaire += macro_f1
micro_f1 = f1_score(all_labels_annot2, all_labels_annot3, average='micro')
micro_binaire += micro_f1
print("Macro F1-score annot2 vs annot3 :", round(macro_f1, 4))
print("Micro F1-score annot2 vs annot3 :", round(micro_f1, 4))

macro_f1 = f1_score(all_labels_annot1, all_labels_annot3, average='macro')
marco_binaire += macro_f1
micro_f1 = f1_score(all_labels_annot1, all_labels_annot3, average='micro')
micro_binaire += micro_f1
print("Macro F1-score: annot1 vs annot3", round(macro_f1, 4))
print("Micro F1-score: annot1 vs annot3", round(micro_f1, 4))

marco_binaire /= 3
print("macro multi",  marco_binaire)
micro_binaire /= 3
print("micro multi",  micro_binaire)
```
